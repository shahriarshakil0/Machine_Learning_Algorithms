```
import pandas as pd
import numpy as np
from sklearn import tree
import pydotplus
from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
```

```
df=pd.read_csv('/content/drive/MyDrive/Data mining & ML/heart_failure.csv')
df
```

Check Null Value

```
df.isnull().values.any()
```

```
df.describe()
```

Seperate Feature Columns

```
x = df.drop(['DEATH_EVENT'], axis = 1)
x
```

Seperate Class Column

```
y = df['DEATH_EVENT']
y
```

```
dummy_y = np_utils.to_categorical(y)
```

```
dummy_y.shape
```

Train, Test & Split our dataset into 70:30

```
xtrain,xtest,ytrain,ytest= train_test_split(x,dummy_y,test_size=.3,random_state=1)
```

Create an ANN model with Kears,
Fit our Train dtata into our model

```
model = Sequential()
model.add(Dense(12, input_shape=(12,), activation='relu')) #input_shape 12 because there are 12 features in x
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='sigmoid')) #output is 2 because we classify 2 types of classes
```

```
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(xtrain, ytrain, batch_size = 5, epochs = 200)
```

```
_, accuracy = model.evaluate(xtest, ytest)
print('Accuracy: %.2f' % (accuracy*100))
```

```
y_pred = model.predict(xtest)
y_pred=np.argmax(y_pred, axis=1)
```

```
y_acctual=np.argmax(ytest, axis=1)
```

```
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_acctual, y_pred)
print(cm)
```

```
print(classification_report(y_acctual, y_pred))
```
