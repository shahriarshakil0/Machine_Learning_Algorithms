# Grids

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(df,hue="transmission")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(df, hue="transmission")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```
